```python
from bs4 import BeautifulSoup as Soup
import requests
from datetime import datetime

def bremen_scrape(baseUrl='https://brementeater.dk/'):

    CONTENT = requests.get(url=baseUrl).text

    soup = Soup(CONTENT, features='lxml')

    concerts = soup.find_all('div', {'class': 'event'})
    concertList = []
    for event in concerts:
        concert = event.find('div', {'class': 'event-title'}).a['title']
        concertUrl = event.find('div', {'class': 'event-title'}).a['href']
        concertDate_ = event.find('div', {'class': 'event-date'}).strong.get_text()
        concertDate = datetime.strptime(concertDate_, '%d.%m.%Y').date()
        ticketUrl = event.find('div', {'class': 'event-ticket-info'}).a['href']
        price = None
        concertList.append([concert, concertUrl, concertDate, ticketUrl, price])
    #print(concertList)
    return concertList


 
if __name__ == '__main__':

    print(bremen_scrape())
```
